```python
def only_even(numbers):

  total = 0
  for number in numbers:
    if number % 2 == 0:
      total += number
  return total

scores = [26, 49, 98, 87, 62, 75]
sum = only_even(scores)
print(f"Sum of even scores: ",sum)
```
